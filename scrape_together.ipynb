```python
import scraper as scr
import pandas as pd

urls = [
    "https://finance.yahoo.com/topic/latest-news/",
    "https://finance.yahoo.com/topic/tech/",
    "https://finance.yahoo.com/topic/housing-market/",
    "https://finance.yahoo.com/topic/stock-market-news/",
    "https://finance.yahoo.com/topic/yahoo-finance-originals/",
    "https://finance.yahoo.com/topic/economic-news/"
    
]

prev_saved_df = pd.read_csv("11_20_26_articles_combined.csv")

dataframes = []

for url in urls:
    print(f"Scraping data from: {url}")
    temp_df = scr.scrape_data(main_url=url)
    dataframes.append(temp_df)



```

```python
dataframes.append(prev_saved_df)
```

```python
# Combine and deduplicate
final_df = scr.combine_and_deduplicate_dataframes(dataframes)

final_df = scr.filter_urls_with_html(final_df)
```

```python
prefix_to_remove = "We are experiencing some temporary issues. The market data on this page is currently delayed. Please bear with us as we address this and restore your personalized lists. "

suffix_to_remove = " Sign in to access your portfolio"

# Remove the prefix from the content column
final_df["content"] = final_df["content"].str.replace(prefix_to_remove, "", regex=False)
final_df["content"] = final_df["content"].str.replace(suffix_to_remove, "", regex=False)
final_df = final_df[final_df["title"] != 'Error']
final_df = final_df.reset_index(drop=True)
```
