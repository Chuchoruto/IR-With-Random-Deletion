```python
import scraper as scr
import pandas as pd

urls = [
    "https://finance.yahoo.com/topic/latest-news/",
    "https://finance.yahoo.com/topic/tech/",
    "https://finance.yahoo.com/topic/housing-market/",
    "https://finance.yahoo.com/topic/stock-market-news/",
    "https://finance.yahoo.com/topic/yahoo-finance-originals/",
    "https://finance.yahoo.com/topic/economic-news/"
    
]

prev_saved_df = pd.read_csv("yahoo_finance_articles_with_content.csv")

dataframes = []

for url in urls:
    print(f"Scraping data from: {url}")
    temp_df = scr.scrape_data(main_url=url)
    dataframes.append(temp_df)



```

```python
# Combine and deduplicate
final_df = scr.combine_and_deduplicate_dataframes(dataframes)

final_df = scr.filter_urls_with_html(final_df)
```
